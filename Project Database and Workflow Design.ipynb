#### What is the trend of people naming their babies in each race? Do baby's name correlate with adult's name? Do people give their baby the same name as their fur baby? 

### Project Description: 

"What makes for a good baby name?" Sometimes people name their babies with their names. Besides, will people name their dog just like naming their human baby? Inspired by the "Big Bang Theory", we decide to make an analysis about the naming trends in US. 

The project analyses two relationships. First, analyze the top 10 baby names in each race and the most popular adult name in each race. Second, analyze the popular dog names and the popular adult name. The goal of the project is to determine if the adult name influences the dog and baby's naming. 

The data sources are based on New York City, and they can demonstrate the trends of the naming habit, respectfully. Within the project, we will sort the top 10 babies with each race from BabyNames.csv from each year. At this part, we will receive some data visualization graphs about the baby names trend. Then, we will use the data source from FirstNames.csv to organize and sort the most common ten names in each race and compare them with the top 10 baby names in each year. As a result, we will receive the correlation percentage of each year with each race. Last, the data source of the dog names has the name and count attribute. We compare the dog names with the adult names again and receive the result of the similarity. In the end, we can find the relationship between dog names, baby names, and adult names with each race. 


### Description of the data sources:


#### 1. DogNames.csv : counts of each dog names

#### 2. FirstNames.csv : 4,250 first names and information on their respective count and proportions 

#### 3. FirstNamesOfGroup.csv : 6 mutually exclusive racial and Hispanic origin groups

#### 4. BabyNames.csv :  7 races represented in New York
